```python
from bs4 import BeautifulSoup
import requests
url=BeautifulSoup('https://www.cashbackforex.com/contests?tid=ggl030823&lang_redirect=1','html.parser')
soup=requests.get(url)
soup
```

```python
raw_code=soup.text
raw_code
```

```python
raw_lxml_code=BeautifulSoup(raw_code,"lxml") #conversion into lxml format
raw_lxml_code
```

```python
table_code=raw_lxml_code.table#extracting table only from source code
 
table_code
```

```python
tags=table_code.find_all('tr')
tags
```

```python
data=[]                  #creation of an empty list
for i in tags:
    x=i.text.strip()
        
    data.append(x.split('\n'))
                          #this helps to split the data from \n and op in list inside list format
    
data    
```

```python
import csv
file=open('fox_data.csv','w')
x=csv.writer(file)
x.writerows(data)
file.close()
```
